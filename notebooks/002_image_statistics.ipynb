```python
import pathlib
import cv2
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm

data_path = pathlib.Path("../data")
train_image_path = data_path / "dtc_train_images"
test_image_path = data_path / "dtc_test_images"
```

```python
mean_img = 0

image_paths = sorted(data_path.glob("*/*.jpg"))
for p in tqdm(image_paths):
    img = cv2.imread(p.as_posix())
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img / 255
    mean_img += img / len(image_paths)
```

```python
plt.imshow(mean_img)
```

```python
cv2.imwrite(
    (data_path/"mean_img.jpg").as_posix(),
    mean_img.dot(255).astype("uint8"),
)
```
